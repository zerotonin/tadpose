```python
def find_missing_numbers(input_str, start=0, end=31):
    """
    Finds missing numbers within a specified range from a comma-separated list.

    Parameters:
    - input_str (str): A string of numbers separated by commas.
    - start (int): The start of the range (inclusive).
    - end (int): The end of the range (inclusive).

    Returns:
    - List[int]: A sorted list of missing numbers.
    """
    # Step 1: Split the input string by commas and convert to integers
    try:
        # Remove any whitespace and convert each split item to integer
        numbers = [int(num.strip()) for num in input_str.split(',')]
    except ValueError:
        raise ValueError("Input must be a comma-separated list of integers.")

    # Step 2: Create a set of all numbers within the desired range
    full_set = set(range(start, end + 1))

    # Step 3: Convert the input list to a set to remove duplicates and for efficient lookup
    given_set = set(numbers)

    # Step 4: Identify missing numbers by set difference
    missing_numbers = sorted(full_set - given_set)

    return missing_numbers

# Example Usage

# Input: Comma-separated list of numbers (as a string)
input_numbers = "0, 2,5,8,9,16,19,20,26,31,1,3,4,6,7,10,12,15,17,18,21,22,24,25,27,28,30,21,33,34,35,13,23,29,11,14"

# Find missing numbers from 1 to 31
missing = find_missing_numbers(input_numbers, 0, 35)

# Display the result
print("Missing numbers from 1 to 31:", missing)
```

```python
def style_dataframe(df, exclude_columns=['Cluster_Label']):
    """
    Styles the DataFrame by coloring cells with '+' as green and '-' as red.
    All other cells remain uncolored. The function excludes specified columns from styling.

    Parameters:
    - df (pd.DataFrame): The DataFrame to style.
    - exclude_columns (list): List of column names to exclude from styling.

    Returns:
    - pd.io.formats.style.Styler: The styled DataFrame.
    """
    
    # Define the cell highlighting function
    def highlight_cells(val):
        if val == '+':
            return 'background-color: lightgreen'
        elif val == '-':
            return 'background-color: lightcoral'
        else:
            return ''
    
    # Identify columns to style by excluding the specified columns
    cols_to_style = [col for col in df.columns if col not in exclude_columns]
    
    # Apply the highlighting function to the selected columns
    styled_df = df.style.applymap(highlight_cells, subset=cols_to_style)
    
    # Enhance table aesthetics: center alignment and borders
    styled_df = styled_df.set_table_styles([
        {'selector': 'th', 
         'props': [('font-size', '12pt'), 
                   ('text-align', 'center'), 
                   ('border', '1px solid black')]},
        {'selector': 'td', 
         'props': [('text-align', 'center'), 
                   ('border', '1px solid black')]}
    ])
    
    return styled_df
```
